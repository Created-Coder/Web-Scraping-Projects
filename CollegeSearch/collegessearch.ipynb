```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
import shutil
from csv import writer
import urllib.request
import csv
import os
```

```python
link = input("Enter Link : ")
filename = input("Enter File Name for saving file, For Example(Data.csv): ")

try:
    driver = webdriver.Chrome(executable_path='chromedriver.exe')
    driver.get(link) 

    with open(os.getcwd() + "/" + filename, 'a+', newline='', encoding='utf-8-sig') as f:
        writer_data = writer(f)
        writer_data.writerow(["College Name", "State", "City", "Course Name", "Duration", "Total Fees", "Institute Type", "Placement - average", "Placement - highest"])

    flag = True
    while(flag):
        try:
            driver.find_element_by_xpath("//li[@class='disabled']/a[@title='Next']")
            flag = False
            driver.close()

        except:
            data = driver.find_elements_by_xpath("//div[@class='well mg-5-0 search_result_card']")
            for i in data:
                try:
                    collegeName = i.find_element_by_xpath(".//h2/a")
                    collegeName = collegeName.get_attribute("title")
                    print("College Name : " + collegeName)

                except:
                    collegeName = " "
                    print("College Name : " + collegeName)

                try:
                    collegeState = i.find_element_by_xpath(".//small")
                    collegeState = collegeState.get_attribute("innerHTML")
                    collegeState = collegeState.split()
                    city = collegeState[0]
                    state = collegeState[1:]
                    state = ' '.join([str(elem) for elem in state])
                    print("College City : " + city)
                    print("College State : " + state)

                except:
                    state = " "
                    city = " "
                    print("College State : " + state)
                    print("College City : " + city)

                try:
                    courseName = i.find_element_by_xpath("//p[@class='course_name']/a")
                    courseName = courseName.get_attribute("innerHTML")
                    print("Course Name : " + courseName)

                except:
                    courseName = " "
                    print("Course Name : " + courseName)

                try:
                    courseDuration = i.find_element_by_xpath(".//*[contains(text(), 'Duration:')]/following-sibling::dd[1]").get_attribute("innerHTML")
                    print("Course Duration : " + courseDuration)

                except:
                    courseDuration = " "
                    print("Course Duration : " + courseDuration)

                try:
                    courseFees = i.find_element_by_xpath(".//*[contains(text(), 'Total Fees:')]/following-sibling::dd[1]").get_attribute("innerHTML")
                    courseFees = courseFees.split(" ")
                    courseFees = courseFees[-1].replace(",", "").replace("/-" , "")
                    print(courseFees)

                except:
                    courseFees = " "
                    print("Course Fees : " + courseFees)

                try:
                    instituteType = i.find_element_by_xpath(".//*[contains(text(), 'Institute Type:')]/following-sibling::dd[1]").get_attribute("innerHTML")
                    print("Institute Type : " + instituteType)

                except:
                    instituteType = " "
                    print("Institute Type : " + instituteType)


                try:
                    placements = i.find_element_by_xpath(".//*[contains(text(), 'Placements:')]/following-sibling::dd[1]").text
                    placements = placements.split("- ")
                    average = placements[1].split()
                    average = average[1]

                    print("Average : " + average)

                    try:
                        highest = placements[3].split()
                        highest = highest[1]
                        print("Highest : " + highest)

                    except:
                        pass

                except:
                    placements = [" "," "," "," "]

                    average = placements[1]
                    highest = placements[3]
                    print("Average : " + average)
                    print("Highest : " + highest)



                fields=[collegeName, state, city, courseName, courseDuration, courseFees, instituteType, average, highest]
                with open(os.getcwd() + "/" + filename, 'a+', newline='', encoding='utf-8-sig') as f:
                    writer_data = writer(f)
                    writer_data.writerow(fields)

                print("--------------------------------")

    #         time.sleep(5)
            wait = WebDriverWait(driver, 20)
            element = wait.until(EC.element_to_be_clickable((By.XPATH, "//a[@title='Next']"))).click()

except:
    pass
    
        
```
