```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.remote.webelement import WebElement
from selenium.webdriver.common.action_chains import ActionChains
import csv

import time
```

```python
# Open Webpage on ChromeDriver


driver = webdriver.Chrome(executable_path='chromedriver.exe')
driver.maximize_window()
driver.get('https://www.wizardingworld.com/quiz/the-hogwarts-express-quiz')
time.sleep(5)
driver.execute_script(f"window.scrollTo(0,1000);")
time.sleep(5)

startBtn = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//h1"))
    ).text
    
# startBtn = driver.find_element_by_xpath("")
print(startBtn)
```

```python
with open("links.csv") as f:
    data = csv.reader(f)
    for i in data:
        for j in i:        
            driver = webdriver.Chrome(executable_path='chromedriver.exe')
            driver.maximize_window()
            driver.get(j)

            
#             driver.execute_script(f"window.scrollTo(0,1200);")
#             time.sleep(2)

            iframe = WebDriverWait(driver, 30).until(
                    EC.presence_of_element_located((By.XPATH, "(//iframe)[1]"))
                )

            print(iframe.get_attribute('src'))
        
            with open('quizLinks.csv', 'a+', encoding='UTF8', newline='') as f:
                writer = csv.writer(f)        
                writer.writerow([iframe.get_attribute('src')])
                print(iframe.get_attribute('src'))

```
