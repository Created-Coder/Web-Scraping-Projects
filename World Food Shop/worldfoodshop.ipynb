```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
import shutil
from csv import writer
import urllib.request
import csv
```

```python
links_list = []
```

```python
with open('G:/World Food Shop/Links_2.csv', 'r') as file:
    reader = csv.reader(file)
    for count, row in enumerate(reader):
        for i in row:
            driver = webdriver.Chrome(executable_path='C:/Users/92324/Downloads/chromedriver_win32 (4)/chromedriver.exe')
            driver.get(i)
            time.sleep(10)
            links = driver.find_elements_by_xpath("//div[@class='flex-col m-col-6 d-col-3 prod-item']//div[@class='prod-item-name-ellipsis']/a")

            for i in links:
                print(i.get_attribute("href"))
                links_list.append(i.get_attribute("href"))

                with open("G:/World Food Shop/links_Remaining.csv", 'a+', newline='') as write_obj:
                        # Create a writer object from csv module
                        csv_writer = writer(write_obj)
                        # Add contents of list as last row in the csv file
                        csv_writer.writerow([i.get_attribute("href")])


            driver.close()
```
