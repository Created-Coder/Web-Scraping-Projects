```python
def get_listing(driver):
    listing = driver.find_elements_by_xpath("//a[@class='annonce']")

    prds = []

    for i in listing:
        prds.append(i.get_attribute("href"))

    return prds
```

```python
def get_prds(driver):
    try:
        driver.implicitly_wait(10)
        prd_title = driver.find_element_by_xpath("//h1[@class='noStickLeft']").text
        
        invalid = '<>:"/\|?*, '

        for i in invalid:
            if i in prd_title:
                prd_title = prd_title.replace(i, " ")
        
        print("title : " + prd_title)
        
        
    except:
        prd_title = " "
        print("title : " + prd_title)
        
        
    try:
        driver.implicitly_wait(10)
        category = driver.find_element_by_xpath("//nav[@id='CDFannonces']//a[2]").text
        print("Category : " + category )
        
    except:
        category = " "
        print("Category : " + category )
        
        
    try:
        driver.implicitly_wait(10)
        sub_category = driver.find_element_by_xpath("//nav[@id='CDFannonces']//a[3]").text
        print("Sub Category : " + sub_category)
        
    except:
        sub_category = " "
        print("Sub Category : " + sub_category)
        
        
    try:
        driver.implicitly_wait(10)
        price = driver.find_element_by_xpath("//*[contains(text(), 'Prix')]/following-sibling::td/strong").text
        print("Price : " + price)
        
    except:
        price = " "
        print("Price : " + price)
        
        
    try:
        driver.implicitly_wait(10)
        address = driver.find_element_by_xpath("//*[contains(text(), 'Lieu')]/following-sibling::td").text
        print("Address : " + address)
        
    except:
        address = " "
        print("Address : " + address)
        
    
    try:
        driver.implicitly_wait(10)
        date = driver.find_element_by_xpath("//*[contains(text(), 'Le')]/following-sibling::td").text
        print("Date : " + date)
        
    except:
        date = " "
        print("Date : " + date)
        
        
    try:
        driver.implicitly_wait(10)
        desc = driver.find_elements_by_xpath("//div[@id='texteAnnonce']")
        desc = '\n '.join([str(elem.text) for elem in desc])
        print("Desc : " + desc)
        
    except:
        desc = " "
        print("Date : " + desc)
        
    try:
        driver.implicitly_wait(10)
        src = driver.find_element_by_xpath("//div[@class='currentPhotoPanel']//img").get_attribute("src")
        
        print("Image URL : " + src)
        
    except:
        src = " "
        print("Image URL : " + src)
        
    try:
        urllib.request.urlretrieve(src, os.getcwd() + "/images/" + prd_title + ".jpg")
    
    except:
        pass
    
    with open(os.getcwd() + "/data_" + str(today) + ".csv", 'a+', newline='', encoding='utf-8') as write_obj:
        csv_writer = writer(write_obj)
        csv_writer.writerow([prd_title, category, sub_category, price, address, date, desc, src, os.getcwd() + "/images/" + prd_title + ".jpg" ])

         
        
    
        
```

```python
today = date.today()

with open(os.getcwd() + "/data_" + str(today) + ".csv", 'a+', newline='', encoding='utf-8') as write_obj:
    csv_writer = writer(write_obj)
    csv_writer.writerow(['Product Title', 'Category', 'Sub Category', 'Price', 'Address', 'Ad Posted Date', 'Description', 'Image URL', 'Image Local Path'])
    
driver = webdriver.Chrome(executable_path="chromedriver.exe")

for i in urls:
    driver.get(i)
    prds = get_listing(driver)

    for i in prds:
        driver.get(i)
        get_prds(driver)    
    
    flag = True
    
    while flag == True:
        try:
            nextPage = driver.find_element_by_xpath("//li[@class='nextPage']")       
            nextPage.click()
            prds = get_listing(driver)
            
            for i in prds:
                driver.get(i)
                get_prds(driver)
            
            
        except:
            flag = False
            continue
    

        
    
```
