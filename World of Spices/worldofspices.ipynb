```python
Links = ["https://www.worldfoodshop.uk/legumes-seeds--beans-88-c.asp",
        "https://www.worldfoodshop.uk/misc-85-c.asp",
"https://www.worldfoodshop.uk/sauces-pastes-pickles-89-c.asp",
"https://www.worldfoodshop.uk/spices-seasonings--herbs-87-c.asp",
"https://www.worldfoodshop.uk/beverages-49-c.asp",
"https://www.worldfoodshop.uk/coconut-products-20-c.asp",
"https://www.worldfoodshop.uk/misc-30-c.asp",
"https://www.worldfoodshop.uk/noodlesvermicellies-6-c.asp",
"https://www.worldfoodshop.uk/oils-and-ghee-19-c.asp",
"https://www.worldfoodshop.uk/rice-and-flours-1-c.asp",
"https://www.worldfoodshop.uk/sauces-pastes-pickles-15-c.asp",
"https://www.worldfoodshop.uk/snacks-44-c.asp",
"https://www.worldfoodshop.uk/spices-and-seasonings-39-c.asp",
"https://www.worldfoodshop.uk/tinned-foods-22-c.asp",
"https://www.worldfoodshop.uk/tinned-fruits-and-veg-43-c.asp",
"https://www.worldfoodshop.uk/sauces-pastes-pickles-44527nproducts15curpage-2-15-c.asp"
]


for i in Links:
    driver = webdriver.Chrome(executable_path='C:/Users/92324/Downloads/chromedriver_win32 (4)/chromedriver.exe')
    driver.get(i)
    time.sleep(10)
    links = driver.find_elements_by_xpath("//div[@class='title-wrapper']//a")

    for i in links:
        print(i.get_attribute("href"))
        links_list.append(i.get_attribute("href"))

        with open("G:/World of Spices/links.csv", 'a+', newline='') as write_obj:
                # Create a writer object from csv module
                csv_writer = writer(write_obj)
                # Add contents of list as last row in the csv file
                csv_writer.writerow([i.get_attribute("href")])


    driver.close()
```

```python
with open('G:/World of Spices/links.csv', 'r') as file:
    reader = csv.reader(file)
    for count, row in enumerate(reader):
        for i in row:
            try:
                driver = webdriver.Chrome(executable_path='C:/Users/92324/Downloads/chromedriver_win32 (4)/chromedriver.exe')
                driver.get(i)
                time.sleep(5)
                
                
                try:
                    product_name = driver.find_element_by_xpath("//h1[@class='product-title product_title entry-title']").text
                    print(product_name)
                    
                    
                except:
                    product_name = "NotFound"
                    pass

                try:
                    description = driver.find_element_by_xpath("//*[@id='tab-description']/p").text
                    print(description)
                
                except:
                    description = "NotFound"
                    pass

                try:
                    price = driver.find_element_by_xpath("//div[@class='price-wrapper']/p").text
                    print(price)
            
                except:
                    price = "NotFound"
                    pass
                
                
                try:
                    src = driver.find_element_by_xpath("//div[@class='flickity-slider']//a").get_attribute("href")
                    urllib.request.urlretrieve(src, "G:/World of Spices/" + product_name + ".jpg")
                
                except:
                    pass
                
                
                fields=[product_name,price,description]

                with open("G:/World of Spices/Data.csv", 'a+', newline='', encoding='utf-8-sig') as f:
                    writer_data = writer(f)
                    writer_data.writerow(fields)



            #     print("data saved successfully")

                driver.close()

            except Exception as e:
                print(e)
                print("Url : " + i)
                with open("G:/World of Spices/Bhand.csv", 'a+', newline='') as write_obj:
                    # Create a writer object from csv module
                    csv_writer = writer(write_obj)
                    # Add contents of list as last row in the csv file
                    csv_writer.writerow([i])

                continue
                continue
                
                
```
