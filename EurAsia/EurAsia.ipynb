```python
def copyFiles():
    files = os.listdir("D:/EurAsia/EurAsia/spiders")
    count = []
    for i in files:
        if i.endswith(".jpg"):
            shutil.move("D:/EurAsia/EurAsia/spiders/" + i, "D:/EurAsia/EurAsia/spiders/Images/" + i)
            print(i + " " + "Moved to Lots Folder")
            count.append(i)

    print(str(len(count)) + " " + "Files Moved")
```
